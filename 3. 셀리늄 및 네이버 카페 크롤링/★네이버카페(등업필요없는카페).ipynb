## 네이버카페 띄우고 자동으로 로그인까지

```python
## 네이버카페 띄우고 자동으로 로그인까지

# 네이버카페(PC) 크롬창 띄우기
driver = webdriver.Chrome("chromedriver.exe") # 크롬 드라이버 로드
driver.get("https://cafe.naver.com/firenze")   # 네이버카페 띄우기
time.sleep(2)

# 로그인창 띄우기
driver.find_element_by_css_selector(".gnb_btn_login").click()
```

```python
# 전체글보기 클릭
# 스크롤 내리기
import time
def scroll_down(driver):
  driver.execute_script("window.scrollTo(0,document.body.scrollHeight)")   # 스크롤 맨밑까지 다운
  time.sleep(0.4)
i = 0
while i < 1:
  scroll_down(driver)     
  i += 1

# 전체글보기 클릭
driver.find_element_by_css_selector("#menuLink0").click()
time.sleep(1)
```

```python
# 검색어 검색

# iframe 접근
driver.switch_to_frame('cafe_main')

# 검색어 입력   
검색어 = "유럽"

query_input = driver.find_element_by_css_selector("#query")
driver.find_element_by_css_selector("#query").click()

for c in 검색어:     
    query_input.send_keys(c)     
    time.sleep(0.3)
    
# 검색 버튼 클릭
driver.find_element_by_css_selector('.btn-search-green').click()    
```

## 글 URL 수집 (이미 했으면 또 할 필요 없음)

```python
url_list = []

for i in range(1, 5):  # i는 페이지 번호
    # query(검색어)가 입력된 url
    url = "https://cafe.naver.com/firenze?iframe_url=/ArticleSearchList.nhn%3Fsearch.clubid=10209062%26search.media=0%26search.searchdate=all%26search.defaultValue=1%26userDisplay=15%26search.option=0%26search.sortBy=date%26search.searchBy=0%26search.query=%C0%AF%B7%B4%26search.viewtype=title%26search.page={}".format(i)
    driver.get(url)
    time.sleep(0.5)
    #페이지 링크로 하면 다른카페 크롤링가능 
    
    # iframe 접근
    driver.switch_to_frame('cafe_main')
    
    # URL 크롤링 시작
    overlays = """.article"""
    article_raw = driver.find_elements_by_css_selector(overlays)               # category
    article_raw
    
    for article in article_raw:
        url = article.get_attribute('href')   
        url_list.append(url)
    
    print(i)

url_list
```

```python
df = pd.DataFrame({'url':url_list})
df
```

```python
# "url_list.csv" 저장하기
df.to_csv("url_list.csv")
```

## 글 크롤링

```python
# "url_list.csv" 불러오기
url_load = pd.read_csv('url_list.csv')        # 기본 모델
url_load
```

# 크롤링하기

```python
dict = {}  # 전체 크롤링 데이터를 담을 그릇
```

```python
for i in tqdm(range(0, len(url_load))):
    try:
        # 글 띄우기
        url = url_load['url'][i]
        driver.get(url)   # 최신글 띄우기
        time.sleep(1.5)
        
        # iframe 접근
        driver.switch_to_frame('cafe_main')

        target_info = {}   # 글 하나의 딕셔너리
    #     # 카테고리 크롤링 시작
    #     overlays = "#main-area .inbox .fl a.m-tcol-c"
    #     cate = driver.find_element_by_css_selector(overlays)      # category
    #     category = cate.text
    #     category

        # 제목 크롤링 시작
        overlays = ".title_text"                                 
        tit = driver.find_element_by_css_selector(overlays)         # title
        title = tit.text
        title
        
        # 내용(content) 크롤링 시작
        overlays = ".ContentRenderer"                                  
        cont = driver.find_element_by_css_selector(overlays)        # content
        content = cont.text
        content

        # 작성자(id) 크롤링 시작
        overlays = ".nickname"                                  
        id_ = driver.find_element_by_css_selector(overlays)         # person
        person = id_.text
        person

        # 날짜(date) 크롤링 시작
        overlays = ".date"                
        date_ = driver.find_element_by_css_selector(overlays)        # datetime
        datetime = date_.text
        datetime

        # 댓글(reply) 크롤링 시작
        overlays = ".text_comment"                                  
        reply = driver.find_elements_by_css_selector(overlays)      # re_list
        r_num = len(reply)
        re_list = []       
        for re in reply:    
            re_list.append(re.text)
        replies = ",".join(re_list)
        replies
        
#         # 이미지(image) 크롤링 시작
#         overlays = "article_img.ATTACH_IMAGE"                                 
#         img_tmp = driver.find_elements_by_name(overlays)         # image

#         img_list = []   
#         for img in img_tmp:    
#             img_url = img.get_attribute('src')
#             img_list.append(img_url)
#         image = ",".join(img_list)
#         image 

        # 글 1개 딕셔너리에 크롤링한 데이터 집어넣기
        #target_info['category'] = category
        target_info['title'] = title
        target_info['url'] = url
        target_info['content'] = content
        target_info['person'] = person
        target_info['datetime'] = datetime
        target_info['reply'] = replies
#         target_info['image'] = image

        # 글 전체 딕셔너리에 글 각각 딕셔너리 넣기
        # target_info

        dict[i] = target_info
        time.sleep(1)

        print(i, title)
    
    except:        
        continue   
```
