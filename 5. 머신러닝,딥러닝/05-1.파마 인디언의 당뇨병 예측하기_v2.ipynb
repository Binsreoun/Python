```python
from keras.models import Sequential
from keras.layers import Dense
import numpy
import tensorflow as tf
```

```python
# seed 값 생성 (랜덤하지 않게 !)
seed = 0
numpy.random.seed(seed)
tf.random.set_seed(seed)    # tensorflow 1.15버전은 tf.set_random_seed(seed)
```

```python
# 데이터 로드
dataset = numpy.loadtxt("./data/pima-indians-diabetes.csv", delimiter=",")
dataset
```

```python
X = dataset[:,0:8]  # 컬럼 0~7까지를 feature로 한다.
Y = dataset[:,8]    # 컬럼 8을 class로 한다.
```

```python
print(X.shape)
X
```

```python
print(Y.shape)
Y
```

```python
# 모델의 설정
model = Sequential()     #바구니
model.add(Dense(16, input_dim=8, activation='relu')) #층
model.add(Dense(8, activation='relu')) #층
model.add(Dense(1, activation='sigmoid')) #층
```

```python
# 모델 컴파일    #이진분석                                     #정확도
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# 모델 실행
model.fit(X, Y, epochs=200, batch_size=10)
```

```python
# 결과 출력
print("\n Accuracy: %.4f" % (model.evaluate(X, Y)[1]))
```
